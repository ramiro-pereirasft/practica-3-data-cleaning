```python
n = int(input("Ingrese un n√∫mero entero"))
for i in range (1, n + 1):
    if i % 3 == 0 and i % 5 == 0:
        print("Tic toc")
    elif i % 3 == 0:
        print("tic")
    elif i % 5 == 0:
        print("toc")
    else :
        print(i)
```

```python
def parcial(number):
    for i in range (1, number + 1):
        if i % 3 ==0 and i % 5==0:
            print("Tic toc")
        elif i % 3 == 0:
            print("tic")
        elif i % 5 == 0:
            print("toc")
        else :
            print(i)


parcial(50)
```
